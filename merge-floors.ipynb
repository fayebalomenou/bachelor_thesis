```python
from pathlib import Path

# Paths
base_path    = Path("/Users/fayebalomenou/Downloads/FINALTHESIS")
input_files  = [
    base_path / "RDF_triples" / "Floor_10" / "floor_10.ttl",
    base_path / "RDF_triples" / "Floor_11" / "floor_11.ttl",
    base_path / "RDF_triples" / "Floor_12" / "floor_12.ttl",
]
output_file = base_path / "merged_floors.ttl"

seen_prefixes = set()
output_lines  = []

for file in input_files:
    print(f"Reading {file.name}...")
    with open(file, "r", encoding="utf-8") as f:
        for line in f:
            stripped = line.strip()
            if stripped.startswith("@prefix"):
                if stripped not in seen_prefixes:
                    seen_prefixes.add(stripped)
                    output_lines.append(line)
            else:
                output_lines.append(line)

with open(output_file, "w", encoding="utf-8") as f:
    f.writelines(output_lines)

print(f"Merged file written → {output_file}")
print(f"Total lines: {len(output_lines)}")
print(f"Unique prefixes: {len(seen_prefixes)}")
```
